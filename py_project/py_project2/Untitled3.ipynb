```python
import numpy as np 
```

```python
t=np.arange(0,12,0.01) #0부터 12까지 0.01 단위로 커짐.
t
```

```python
y=np.sin(t)
y
```

```python
plt.figure(figsize=(10,6))
plt.plot(t,y)
plt.grid()#눈금선
plt.show()
```

```python
plt.figure(figsize=(10,6)) #그래프 크기 가로 :6 세로  :10
plt.plot(t,y)
plt.grid()#눈금선
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Example of sinewawe")
plt.show()
```

```python
plt.figure(figsize=(10,6)) #그래프 크기 가로 :6 세로  :10
plt.plot(t,np.sin(t),lw=3, label='sin')
plt.plot(t,np.cos(t),'r', label='cos') #r- red
plt.grid()#눈금선
plt.legend()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Example of sinewawe")
plt.show()
```

```python
plt.figure(figsize=(10,6)) #그래프 크기 가로 :6 세로  :10
plt.plot(t,np.sin(t),lw=3, label='sin')
plt.plot(t,np.cos(t),'r', label='cos') #r- red
plt.grid()#눈금선
plt.legend()
plt.xlabel("time")
plt.ylabel("Amplitude")
plt.title("Example of sinewawe")
plt.ylim(-1.2, 1.2)  #전체영역에서 보고싶은 것만 보고싶은 경우
plt.xlim(0,np.pi)
plt.show()
```

```python
s1=np.random.normal(loc=0,scale=1,size=1000) #위치가 0을 중심으로, scale=1인, normal분포
s2=np.random.normal(loc=5,scale=0.5,size=1000)
s3=np.random.normal(loc=10,scale=2,size=1000)
```

```python
plt.figure(figsize=(10,6))
plt.plot(s1,label='s1')
plt.plot(s2,label='s2')
plt.plot(s3,label='s3')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.boxplot((s1,s2,s3)) #노랑색 -> 중앙 값  / 동그라미 ->이상치 (아웃라이어) / - -> 최소,25%,중앙값(50%),75%,최대값으로 구성
plt.grid()
plt.show()
```
